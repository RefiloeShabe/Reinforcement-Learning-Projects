```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
# Creating a directory for plots
os.makedirs('qtables')
```

```python
# Defining the gym environment we will be training
env = gym.make("MountainCar-v0")

#-----Paramenters of the model-----------
LEARNING_RATE = 0.1
DISCOUNT = 0.95
EPISODES = 25000
SHOW_EVERY = 3000
STATS_EVERY = 100

DISCRETE_OS_SIZE = [20] * len(env.observation_space.high)
discrete_os_win_size = (env.observation_space.high - env.observation_space.low) / DISCRETE_OS_SIZE

# EXPLORATION SETTINGS
# We want to allow the agent to randomly explore the environment even after figuring out how to solve the problem
epsilon = 0.5
START_EPSILON_DECAYING = 1
END_EPSILON_DECAYING = EPISODES 
epsilon_decay_value = epsilon / (END_EPSILON_DECAYING - START_EPSILON_DECAYING)

# The q table: 
'''a uniform distribution of shape (20,20,3), 
Combination of 20 possible observation by 3 actions 
with q values between (-2,0) that neeed to be optimised over time'''
q_table = np.random.uniform(low = -2, high = 0, size =(DISCRETE_OS_SIZE + [env.action_space.n]))

# An agent gets a reward for every action it takes and it aims to maximize this reward over time 
ep_rewards = []

# Dictionary of episode number, average reward, minimum reward and maximum reward
aggr_ep_rewards = {'ep': [], 'avg': [], 'min': [], 'max': []}


def get_discrete_state(state):
    '''This is a helper function that coverts the continuous states into discrete states '''
    discrete_state = (state - env.observation_space.low) / discrete_os_win_size
    return tuple(discrete_state.astype(np.int))


for episode in range(EPISODES):
    episode_reward = 0
    if episode % SHOW_EVERY == 0:
        print(episode)
        render = True
      
    else:
        render = False
        
    discrete_state = get_discrete_state(env.reset())
    done = False
    while not done:
        
        if np.random.random() > epsilon:
            # get the an action that gives maximum reward (Q learning)
            action = np.argmax(q_table[discrete_state])
            
        else:
            # Get a random action
            action = np.random.randint(0, env.action_space.n)
            
            
        new_state, reward, done, _ = env.step(action)
        
        episode_reward += reward
        
        new_discrete_state = get_discrete_state(new_state)
        
        #if render:
        if episode % SHOW_EVERY == 0:
            env.render()
        
        # If the agent is not done learning the environment, calculate the new Q-value
        if not done:
            
            # Maximum possible Q value for new state
            max_future_q = np.max(q_table[new_discrete_state])
            
            # Current Q value for current state wrt action performed
            current_q = q_table[discrete_state + (action, )]
            
            # New Q value for current state wrt action performed 
            new_q = (1 - LEARNING_RATE) * current_q + LEARNING_RATE * (reward + DISCOUNT * max_future_q)
            
            # update the q table with the new q values
            q_table[discrete_state + (action, )] = new_q
        
        # If the environment is complete, there is no reward given so the q table is updated with a 0
        elif new_state[0] >= env.goal_position:
            #print(f"We made it on episode {episode}")
            q_table[discrete_state + (action, )] = 0
            
        discrete_state = new_discrete_state 
         
    
    # Decaying at every episode if episode number within decaying range
    if END_EPSILON_DECAYING >= episode >= START_EPSILON_DECAYING:
        epsilon -= epsilon_decay_value
    
    ep_rewards.append(episode_reward)
    
    if not episode % STATS_EVERY:
    
        #average_reward = sum(ep_rewards[-SHOW_EVERY:]) / len(ep_rewards[-SHOW_EVERY:])
        average_reward = sum(ep_rewards[-STATS_EVERY:]) / STATS_EVERY
        aggr_ep_rewards['ep'].append(episode)
        aggr_ep_rewards['avg'].append(average_reward)
        aggr_ep_rewards['min'].append(min(ep_rewards[-STATS_EVERY:]))
        aggr_ep_rewards['max'].append(max(ep_rewards[-STATS_EVERY:]))
        
        print(f"Episode: {episode} avg: {average_reward}, current epsilon: {epsilon} ") # min: {min(ep_rewards[-SHOW_EVERY:])} max: {max(ep_rewards[-SHOW_EVERY:])}")
        
        if episode % 100 == 0:
            np.save(f"qtables/{episode}-qtable.npy", q_table)
        
        
env.close()

fig = plt.figure(figsize = (12,9))
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['avg'], label = "avg")
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['min'], label = "min")
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['max'], label = "max")
plt.xlim(3000,25000)
plt.ylim(-200, -100)
plt.legend(loc = 4)
plt.show()
```
